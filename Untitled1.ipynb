## DISCOUNT AND PROFIT AND SALES

```python
plt.figure(figsize=(15,10))
sns.lineplot('Discount', 'Profit', data= data, color='g',label='Discount')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot('Discount', 'Sales', data= data, color='g',label='Discount')
plt.legend()
```

From the lineplot of sales vs discount we can conclude that there is a negative association between these two variables i.e if the value of discount increases, the value of profit decrease.The value of sales is highest when discount increases, the value of profit decreases.The value of sales is highest when discount ig given at rate 50% then the amount of sales decreases drastically

## DISCOUNT AND QUANTITY

## SALES PROFIT PER CATEGORY 

```python
sales_profit = data.groupby(['Category'])['Profit','Sales'].agg(['sum'])
sales_profit.plot.bar(figsize=(15,12))
```

## DISCOUNT PROFIT PER SUB-CATEGORY 

```python
discount_profit = data.groupby(['Category'])['Profit','Sales'].agg(['sum'])
discount_profit.plot.bar(figsize=(15,12))
```

## DISCOUNT PROFIT PER SEGMENT 

```python
discount_profit = data.groupby(['Segment'])['Profit','Discount'].agg(['sum'])
discount_profit.plot.bar(figsize=(15,12))
```
